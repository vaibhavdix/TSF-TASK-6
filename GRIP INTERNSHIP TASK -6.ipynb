###### A DECISION TREE is a non-parametric supervised learning algorithm, which is utilized for both classification and regression tasks. It has a hierarchical, tree structure, which consists of a root node, branches, internal nodes and leaf nodes

## STEP 1 : LOAD DATASET AND REQUIRED LIBRARIES

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
data=pd.DataFrame(iris['data'],columns=["petal_lenth","petal_width","sepal_lenth","sepal_width"])
```

```python
data['species']=iris['target']
```

```python
data.head()
```

## STEP 2 :- UNDERSANDING DATASET

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

## STEP 3 :- VISUALISING DATASET

```python
data.corr()
```

```python
nums={'species':{'iris-setosa':0,'iris-versicolor':1,'iris-virginica':2}}
data.replace(nums,inplace=True)
data.head()
```

## STEP 4 :- BUILDING THE DECISION TREE

```python
x=(data.drop('species',axis=1)).values
y=data['species']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
DT=DecisionTreeClassifier(random_state=12)
model=DT.fit(x_train,y_train)
```

```python
DT.predict(x_test)
```

## STEP 5 :- TESTING THE ACCURACY OF MODEL

```python
y_pred=DT.predict(x_test)
```

```python
print('Accuracy Score:',metrics.accuracy_score(y_test,y_pred))
```

## STEP 6 :- BUILDING THE TREE VISUALLY

```python
plt.figure(figsize=(20,10))
tree.plot_tree(model,rounded=True,filled=True)
plt.show()
```
